```python
import urllib2
import pandas as pd
import numpy as np
import re

urls = ['https://www.census.gov/did/www/saipe/downloads/estmod93/est93ALL.dat',
        'https://www.census.gov/did/www/saipe/downloads/estmod95/est95ALL.dat',
        'https://www.census.gov/did/www/saipe/downloads/estmod97/est97ALL.dat',
        'https://www.census.gov/did/www/saipe/downloads/estmod98/est98ALL.dat',
        'https://www.census.gov/did/www/saipe/downloads/estmod99/est99ALL.dat',
        'https://www.census.gov/did/www/saipe/downloads/estmod00/est00ALL.dat',
        'https://www.census.gov/did/www/saipe/downloads/estmod01/est01ALL.dat',
        'https://www.census.gov/did/www/saipe/downloads/estmod02/est02ALL.dat',
        'https://www.census.gov/did/www/saipe/downloads/estmod03/est03ALL.dat',
        'https://www.census.gov/did/www/saipe/downloads/estmod04/est04ALL.txt',
        'https://www.census.gov/did/www/saipe/downloads/estmod05/est05ALL.txt',
        'https://www.census.gov/did/www/saipe/downloads/estmod06/est06ALL.txt',
        'https://www.census.gov/did/www/saipe/downloads/estmod07/est07ALL.txt',
        'https://www.census.gov/did/www/saipe/downloads/estmod08/est08ALL.txt',
        'https://www.census.gov/did/www/saipe/downloads/estmod09/est09ALL.txt',
        'https://www.census.gov/did/www/saipe/downloads/estmod10/est10ALL.txt',
        'https://www.census.gov/did/www/saipe/downloads/estmod11/est11all.txt',
        'https://www.census.gov/did/www/saipe/downloads/estmod12/est12ALL.txt',
        'https://www.census.gov/did/www/saipe/downloads/estmod13/est13ALL.txt']

#define function to read web data into python
def getUrl(urls):
    line_list=[]
    for url in urls:
        response = urllib2.urlopen(url)
        lines = response.read().split('\n')
        del lines[-1]
        print url
        line_list.append(lines)
        #TODO -code to use the text data
        #line_list.append(lines)
    return line_list
data=getUrl(urls)

#extract the year information from urls
years=[]
list=range(19)
for i in list:
    s=urls[i]
    cha = s[47:55]
    years.append(cha)

#join year information into data
data1=[]
datayear_len=[]
for k in list:
    data_year=data[k]
    datayear_len.append(len(data_year))
    for line in data_year:
        line = years[k] + ' ' + line
        data1.append(line)

#sepearte data into national, state and county level's data
listus=[]
listst=[]
listcu=[]
for line in data1:
    if "US" in line:              #select country level data
        line = line.split('\n')
        listus.append(line)
    #elif ("County" in line and line[14] != '0'),does not work AK no county word
    elif int(line[11:16])>0:       #select county level data
        line = line.split('\n')  
        listcu.append(line)
    else:                          #state level data 
        line = line.split('\n')
        listst.append(line)
print type(listst), len(listst)
print type(listus), len(listus)
print type(listcu), len(listcu)

#write data into file
datafile1 = open('listus.txt', 'w')
for items1 in listus:
    datafile1.write("%s\n" % items1)

datafile2 = open('listst.txt', 'w')
for items2 in listst:
    datafile2.write("%s\n" % items2)

datafile3 = open('listcu.txt', 'w')
for items3 in listcu:
    datafile3.write("%s\n" % items3)
```

```python
import json
import vincent
vincent.initialize_notebook()

# 1. Overall percentage of poverty for each county in 2000.

county_topo=r'us_counties.topo.json'
geo_data = [{'name': 'counties',
             'url': county_topo,
             'feature': 'us_counties.geo'}]
vis=vincent.Map(data=coupovallages2000,geo_data=geo_data,scale=1100, 
                projection='albersUsa', data_bind='Poverty_2000',
                data_key='FIPS', map_key={'counties': 'properties.FIPS'})
vis.scales['color'].type = 'threshold'
vis.scales['color'].domain = [0, 4, 6, 8, 10, 12, 20, 30]
vis.legend(title='Poverty 2000(%)')
vis.display()
```

```python
# 2. Overall percentage of poverty for each county in 2013.

county_topo=r'us_counties.topo.json'
geo_data = [{'name': 'counties',
             'url': county_topo,
             'feature': 'us_counties.geo'}]
vis=vincent.Map(data=coupovallages2013,geo_data=geo_data,scale=1100, 
                projection='albersUsa', data_bind='Poverty_2013',
                data_key='FIPS', map_key={'counties': 'properties.FIPS'})
vis.scales['color'].type = 'threshold'
vis.scales['color'].domain = [0, 4, 6, 8, 10, 12, 20, 30]
vis.legend(title='Poverty 2013(%)')
vis.display()
```

```python
# 3. Ratio between percentage of change in poverty from 2000 to 2013 for each county. Divided by the nation-wide change in poverty (1.39).

county_topo=r'us_counties.topo.json'
geo_data = [{'name': 'counties',
             'url': county_topo,
             'feature': 'us_counties.geo'}]
vis=vincent.Map(data=coupovallageschange,geo_data=geo_data,scale=1100, 
                projection='albersUsa', data_bind='Poverty_change',
                data_key='FIPS', map_key={'counties': 'properties.FIPS'})
vis.scales['color'].type = 'threshold'
vis.scales['color'].domain = [0, 0.8, 0.9, 1, 1.1, 1.2, 1.3]
vis.legend(title='Change in Poverty')
vis.display()
```
